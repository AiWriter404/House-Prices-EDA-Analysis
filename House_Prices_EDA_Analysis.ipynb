# House Prices EDA Analysis

This notebook is created for practicing **Exploratory Data Analysis (EDA)** using Python libraries.

## Libraries Used
- pandas
- numpy
- matplotlib
- seaborn

## Goals
- Load a dataset
- Explore data structure
- Handle missing values
- Visualize relationships
- Understand patterns in house prices


```python
# Import required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Display plots inside notebook
%matplotlib inline

# Set seaborn style
sns.set_style("whitegrid")

print("Libraries imported successfully!")
```

## Create a House Prices Dataset

Seaborn does not provide a built-in house prices dataset.

So here we create a realistic practice dataset manually using pandas and numpy.


```python
# Create a sample house prices dataset

np.random.seed(42)

data = {
    "Square_Feet": np.random.randint(800, 4000, 200),
    "Bedrooms": np.random.randint(1, 6, 200),
    "Bathrooms": np.random.randint(1, 5, 200),
    "Age": np.random.randint(1, 40, 200),
    "Garage_Size": np.random.randint(0, 3, 200),
}

# Create dataframe
df = pd.DataFrame(data)

# Create house price column
df["Price"] = (
    df["Square_Feet"] * 120 +
    df["Bedrooms"] * 15000 +
    df["Bathrooms"] * 10000 -
    df["Age"] * 2000 +
    df["Garage_Size"] * 8000 +
    np.random.randint(-20000, 20000, 200)
)

# Show first 5 rows
df.head()
```

## Basic Dataset Information

```python
# Check dataset shape

print("Rows and Columns:", df.shape)

# Show column names
print("\nColumns:")
print(df.columns)

# Dataset information
print("\nDataset Information:")
df.info()
```

## Statistical Summary

```python
# Statistical summary

df.describe()
```

## Check Missing Values

```python
# Check missing values

df.isnull().sum()
```

## Correlation Analysis

```python
# Correlation matrix

correlation = df.corr()

correlation
```

## Correlation Heatmap

```python
# Plot correlation heatmap

plt.figure(figsize=(10,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

## House Price Distribution

```python
# Distribution of house prices

plt.figure(figsize=(8,5))

sns.histplot(df["Price"], kde=True)

plt.title("House Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

## Relationship Between Square Feet and Price

```python
# Scatterplot

plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Square_Feet",
    y="Price",
    data=df
)

plt.title("Square Feet vs Price")

plt.show()
```

## Boxplot of Bedrooms and Price

```python
# Boxplot

plt.figure(figsize=(8,5))

sns.boxplot(
    x="Bedrooms",
    y="Price",
    data=df
)

plt.title("Bedrooms vs Price")

plt.show()
```

## Pairplot

```python
# Pairplot for multiple relationships

sns.pairplot(df)

plt.show()
```

# Final Insights

## Important Observations
1. Larger houses usually have higher prices.
2. More bedrooms and bathrooms increase house prices.
3. Older houses may have lower prices.
4. Heatmap helps identify strong correlations.
