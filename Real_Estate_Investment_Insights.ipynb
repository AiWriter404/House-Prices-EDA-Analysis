```python

# Import required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Machine learning libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# Show plots inside notebook
%matplotlib inline

# Set style
sns.set_style("whitegrid")

print("Libraries imported successfully!")
```


## Create Realistic Housing Dataset


```python

# Create sample dataset

np.random.seed(42)

areas = [
    "Downtown",
    "Green Valley",
    "Sunset Town",
    "Lake View",
    "Royal City",
    "Hill Park"
]

data = {
    "Area": np.random.choice(areas, 300),
    "Square_Feet": np.random.randint(700, 4500, 300),
    "Bedrooms": np.random.randint(1, 6, 300),
    "Bathrooms": np.random.randint(1, 5, 300),
    "Garage": np.random.randint(0, 3, 300),
    "Age": np.random.randint(1, 35, 300),
}

df = pd.DataFrame(data)

# Create realistic house prices

df["Price"] = (
    df["Square_Feet"] * 130 +
    df["Bedrooms"] * 18000 +
    df["Bathrooms"] * 12000 +
    df["Garage"] * 9000 -
    df["Age"] * 1500 +
    np.random.randint(-25000, 25000, 300)
)

df.head()
```


## Area Wise Investment Analysis


```python

# Average price by area

area_prices = df.groupby("Area")["Price"].mean().sort_values()

print(area_prices)
```

```python

# Visualize average prices

plt.figure(figsize=(10,5))

area_prices.plot(kind="bar")

plt.title("Average House Prices by Area")
plt.ylabel("Average Price")

plt.show()
```


## Correlation Analysis


```python

# Correlation heatmap

correlation = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```
